```python
# Given `arr = np.random.randint(1, 101, 20)`:
# Print its sum, min, and max.
# Step 1: Create array
import numpy as np

# Step 1: Create array
arr = np.random.randint(1, 101, 20)
print("Array:", arr)

# Step 2: Sum, min, max
print("Sum:", np.sum(arr))
print("Minimum:", np.min(arr))
print("Maximum:", np.max(arr))

# Step 3: Second array
arr2 = np.random.randint(1, 101, 20)
print("Second Array:", arr2)

# Step 4: Element-wise multiplication
result = arr * arr2
print("Element-wise multiplication:", result)
```

```python
# Create a 5×3 DataFrame of random integers (1–100) with columns `A, B, C`. Print the **sum** and **mean** of each column.
import numpy as np
import pandas as pd

# Create DataFrame
data = np.random.randint(1, 101, (5, 3))
df = pd.DataFrame(data, columns=['A', 'B', 'C'])

print("DataFrame:\n", df)

# Sum of each column
print("\nColumn-wise Sum:\n", df.sum())

# Mean of each column
print("\nColumn-wise Mean:\n", df.mean())
```
